```python
class Vehicle: #Vehicle class is content whose container is Employee class. Aggregation or Composition relationship between the two
    def __init__(self,make,model,year,mileage): #Initializing the attributes of the class Vehicle
        self.__make = make
        self.__model = model
        self.__year = year
        self.__mileage = mileage
        
    #Getters and setter methods
    def get_make(self):
        return self.__make
    
    def get_model(self):
        return self.__model
    
    def get_year(self):
        return self.__year
    
    def get_mileage(self):
        return self.__mileage
    
    def set_make(self,make):
        self.__make = make
    
    def set_model(self,model):
        self.__model = model
    
    def set_year(self,year):
        self.__year = year
    
    def set_mileage(self,mileage):
        self.__mileage = mileage
        
    def __str__(self):
        return "Make: " + str(self.__make) + "; Model: " + str(self.__model) + "; Year: " + str(self.__year) + "; Mileage: " + str(self.__mileage)
```

```python
class Employee: #Employee class
    def __init__(self,emp_name,emp_address,emp_vehdata): #Initializing the attributes of the class Employee and self represents the instance or object of the class
        self.__emp_name=emp_name
        self.__emp_address=emp_address
        self.__emp_vehdata=emp_vehdata
    
    def get_name(self): #Getter method for employee name
        return self.__emp_name

    def get_address(self): #Getter method for employee address
        return self.__emp_address
    
    def get_vehdata(self): #Getter method for employee vehicle data
        return self.__emp_vehdata
    
    def set_name(self, emp_name): #Setter method for employee name
        self.__emp_name = emp_name

    def set_address(self, emp_address): #Setter method for employee address
        self.__emp_address = emp_address
        
    def set_vehdata(self,emp_vehdata): #Setter method for employee vehicle data
        self.__emp_vehdata = emp_vehdata
        
    def get_data_option4(self):
        if self.__emp_vehdata.get_mileage()>78000:
            print("Employee name is : {0} {1}".format(self.get_name(),self.get_vehdata()))
        
    def __str__(self):
        allData = "Employee Name: " + str(self.__emp_name) + "; Employee Address: " + str(self.__emp_address)
        allData +=  "\n" + self.__emp_vehdata.__str__()
        return allData
```

```python
class FullTimeEmployee(Employee): #FullTimeEmployee is child class inheriting from it's parent class Employee
    def __init__(self,emp_name,emp_address,emp_vehdata,salary): #Initializing the attributes of the class FullTimeEmployee
        super().__init__(emp_name,emp_address,emp_vehdata) #Inheriting the attributes of the parent class
        self.__salary=salary
        
    def get_salary(self): #Getter method for employee salary
        return self.__salary
    
    def set_salary(self, salary): #Setter method for employee salary
        self.__salary = salary
        
    def compute_compensation(self): #Compute compensation method for FullTimeEmployee
        sal=self.__salary 
        if sal <= 45000 : 
            comp = sal - (0.18*sal)
            
        elif sal>45000 and sal<=82000:
            comp = sal - (0.18*45000) - ((sal-45000)*0.28)
        
        elif sal>82000 : 
            comp = sal - (0.18*45000) - (0.28*37000) - ((sal-82000)*0.33)
        
        return "{0}'s compensation is ${1:.2f}".format(self.get_name(),comp)
    
    def __str__(self):
        print("\nDetails of Full Time Employee are:")
        childdata = super().__str__()
        childdata += "\nSalary: " + str(format(self.__salary,'.2f'))
        return childdata

    
        
    
        
        
```

```python
class HourlyEmployee(Employee): #HourlyEmployee is child class inheriting from it's parent class Employee
    def __init__(self,emp_name,emp_address,emp_vehdata,hours_worked,hourly_rate): #Initializing the attributes of the class HourlyEmployee
        super().__init__(emp_name,emp_address,emp_vehdata) #Inheriting the attributes of the parent class
        self.__hours_worked=hours_worked
        self.__hourly_rate=hourly_rate
        
    def get_hours_worked(self): #Getter method for employee hours worked
        return self.__hours_worked
    
    def set_hours_worked(self, hours_worked): #Setter method for employee hours worked
        self.__hours_worked = hours_worked
        
    def get_hourly_rate(self): #Getter method for employee hourly rate
        return self.__hourly_rate
    
    def set_hourly_rate(self, hourly_rate): #Setter method for employee hours worked
        self.__hourly_rate = hourly_rate
        
    def compute_compensation(self): #Compute compensation method for HourlyEmployee
        hrsworked=self.__hours_worked
        hrlyrate=self.__hourly_rate
        if hrsworked <= 40: 
            comp = hrsworked*hrlyrate
        else: 
            comp = (40*hrlyrate) + ((hrsworked-40)*hrlyrate*1.8)
        
        return "{0}'s compensation is ${1:.2f}".format(self.get_name(),comp)
    
    def __str__(self):
        print("\nDetails of Hourly Employee are:")
        childdata = super().__str__()
        childdata += "\nHour Rate: " + str(self.__hourly_rate) + "\nHour Worked: " + str(self.__hours_worked)
        return childdata
    
        
```

```python
class Consultant(Employee): #Consultant is child class inheriting from it's parent class Employee
    def __init__(self,emp_name,emp_address,emp_vehdata,hours_worked,project_type): #Initializing the attributes of the class Consultant
        super().__init__(emp_name,emp_address,emp_vehdata) #Inheriting the attributes of the parent class
        self.__hours_worked=hours_worked
        self.__project_type=project_type
        
    def get_hours_worked(self): #Getter method for employee hours worked
        return self.__hours_worked
    
    def set_hours_worked(self, hours_worked): #Setter method for employee hours worked
        self.__hours_worked = hours_worked
        
    def get_project_type(self): #Getter method for employee project type
        return self.__project_type
    
    def set_project_type(self, project_type): #Setter method for employee project type
        self.__project_type = project_type
        
    def compute_compensation(self): #Compute compensation method for Consultant
        project_type=self.__project_type
        hrsworked=self.__hours_worked
        if project_type == 1:
            comp = 55*hrsworked
        
        elif project_type == 2:
            comp = 70*hrsworked
        
        elif project_type == 3:
            comp = 85*hrsworked
                    
        return "{0}'s compensation is ${1:.2f}".format(self.get_name(),comp)
    
    def __str__(self):
        print("\nDetails of Consultant are:")
        childdata = super().__str__()
        childdata += "\nProject Type: " + str(self.__project_type) + "\nHours Worked: " + str(self.__hours_worked)
        return childdata

            
```

```python

import pickle
import sys

def main(): 
    run_menu_options()

def run_menu_options():   
    employee_list = read_data()
    sel = 1
    
    while sel>=1 and sel<=5:
        print("\n==== Menu ====")
        print("1. To add an Employee")
        print("2. To print the name and address of all employees")
        print("3. To print the employee name and compensation of all employees")
        print("4. To print the employee name with vehicle information for high mileage vehicles")
        print("5. TO exit program")

        try:
            sel = int(input("Your selection: "))
            if sel < 1 or sel > 5:
                raise ValueError
        except ValueError:
            print('You must enter an integer between 1 and 5!')
            print('Try again')
            sel = 1
            continue

        if sel == 1:
            employee_list.append(run_option1())
        elif sel == 2:
            if len(employee_list) == 0: 
                print("No employees in database!")
            else:
                run_option2(employee_list) 
        elif sel == 3:
            if len(employee_list) == 0:
                print("No employees in database!")
            else:    
                run_option3(employee_list)
        elif sel == 4:
            if len(employee_list) == 0:
                print("No employees in database!")
            else:    
                run_option4(employee_list)
        elif sel == 5:
            run_option5(employee_list) 

#load data from .dat file
def read_data():  
    my_lst = []
    try:
        el = open('empdata.dat','rb')
    except FileNotFoundError:
        return my_lst
    else : 
        try:
            while(True):
                my_lst.append(pickle.load(el))
        except EOFError:
            el.close()
            return my_lst

#adding employee
def run_option1():
    sel = 0 
    while sel<1 or sel>4:
        try:
            sel = int(input("Type of Employee? (1. Full Time 2. Hourly 3. Consultant)"))
        except ValueError:
            print("Wrong input")
            continue
        
    emp_name,emp_address = get_employee_input()
    make,model,year,mileage = get_vehicle_input()
    ob_veh = Vehicle(make,model,year,mileage)
    
    if sel==1:
        salary = get_fulltime_input()
        ob_emp = FullTimeEmployee(emp_name,emp_address,ob_veh,salary)
    elif sel==2:
        hourly_rate,hours_worked = get_hourly_input()
        ob_emp = HourlyEmployee(emp_name,emp_address,ob_veh,hours_worked,hourly_rate)
    elif sel==3:
        project_type,hours_worked = get_consultant_input()
        ob_emp = Consultant(emp_name,emp_address,ob_veh,hours_worked,project_type)
        
    return ob_emp

#display employee details
def run_option2(my_lst):
    
    for obj in my_lst :
        print(obj)

#display employee name and compensation
def run_option3(my_lst):
    print("\n Employees Name and Compensation of all employees")
    print("===================================================")
    for obj in my_lst:
        print(obj.compute_compensation())

#display employee name and vehicle data
def run_option4(my_lst):
    print("\n Employees Name and Vehicle data for high mileage vehicles")
    print("=============================================================")
    for obj in my_lst:
        obj.get_data_option4()

#exit the program
def run_option5(my_lst):
    print("You chose to exit the program")
    yes_no = input("Are you sure (Y/N)? ")

    if yes_no.lower() == 'y':
        el = open('empdata.dat', 'wb')
        for obj in my_lst:
            pickle.dump(obj, el)
        el.close()
        sys.exit()
        
def get_employee_input():
    
    emp_name = input("Enter Employee's name : ")
    emp_address = input("Enter Employee's address : ")
    
    return emp_name,emp_address

def get_vehicle_input():
    make = input("Enter vehicle's make : ")
    model = input("Enter vehicle's model : ")
    
    err_flag = True
    while err_flag:
        try:
            year = int(input("Enter vehicle's year : "))
        except ValueError:
            print('\nEnter a numerical value for year!!')
        else:
            err_flag = False
    
    err_flag = True
    while err_flag :
        try:
            mileage = int(input("Enter vehicle's mileage : "))
        except ValueError:
            print('\nEnter a numerical value for mileage!!')
        else:
            err_flag = False
    return (make,model,year,mileage)

def get_fulltime_input():
    err_flag = True
    while err_flag:
        try:
            salary = int(input("Enter salary for Full-time employee : "))
        except ValueError: 
            print('\nEnter a numerical value for salary!!')
        else:
            err_flag = False
    return salary
    
def get_hourly_input():
    err_flag = True
    while err_flag:
        try:
            hourly_rate = int(input("Enter rate for Hourly employee : "))
        except ValueError: 
            print('\nEnter a numerical value for hourly rate!!')
        else:
            err_flag = False
    
    err_flag = True
    while err_flag:
        try:
            hours_worked = int(input("Enter hours worked for Hourly employee : "))
        except ValueError: 
            print('\nEnter a numerical value for hours worked!!')
        else:
            err_flag = False
    return (hourly_rate,hours_worked)

def get_consultant_input():
    err_flag = True
    while err_flag:
        try:
            project_type = int(input("Enter project type for consultant : "))
            if project_type>3 or project_type<1:
                raise ValueError
                
        except ValueError: 
            print('\nEnter a numerical value for project type that is between 1 and 3')
        else:
            err_flag = False
    
    err_flag = True
    while err_flag:
        try:
            hours_worked = int(input("Enter hours worked for consultant : "))
        except ValueError: 
            print('\nEnter a numerical value for hours worked!!')
        else:
            err_flag = False
    return (project_type,hours_worked)

main()
```
