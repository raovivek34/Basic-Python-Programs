```python
import math
import matplotlib.pyplot as plot

filedata=open('carprices.txt')
filelines=filedata.readlines()
prices=[]
for line in filelines[1:]:
    try:
        prices.append(math.floor(int(line)/10))        
    except ValueError as e:
        print("This data is invalid and will be ignored.",e)
        
range=(0,45)
bins = 50
plot.hist(prices,bins,range,color='cornflowerblue',histtype='bar',rwidth=0.8)
plot.xlabel("price in tens of thousands")
plot.ylabel("Number of cars")
plot.title("Frequency of Car Prices")
plot.show()
filedata.close()
```

```python
import matplotlib.pyplot as plot
sort1 = []
sort2 = []
small = 0
midsize = 0
compact = 0
large = 0
sporty = 0
van = 0
file2data = open('cartype.txt')
file2lines = file2data.readlines()
for i in file2lines:
    j = i.strip('\n')
    if  j not in sort1:
        if j == 'Type':
            continue
        sort1.append(j)
    sort2.append(j)
    small = sort2.count('small')
    midsize = sort2.count('midsize')
    compact = sort2.count('compact')
    large = sort2.count('large')
    sporty = sort2.count('sporty')
    van = sort2.count('van')
Frequency = [sporty,van,compact,large,midsize,small]
Labels=['sporty','van','compact','large','midsize','small']
Colors=['c','b','m','c','b','m']

plot.pie(Frequency, labels = Labels, colors=Colors, explode = (0.02, 0.02, 0.02, 0.02, 0.03, 0.03),
        startangle=90,radius = 1.2, autopct = '%.2f%%')
plot.show
file2data.close()
```

```python
import matplotlib.pyplot as plot
years=[]
bears=[]
dolphins=[]
whales=[]
file3data=open('Wildlife.txt')
file3lines=file3data.readlines()
for i in file3lines:
    m=i.split('\t')
    try:
        years.append(int(m[0]))
        bears.append(int(m[1]))
        dolphins.append(int(m[3]))
        whales.append(int(m[5]))
    except Exception:
        pass
    
plot.plot(years,bears,label='Bears')
plot.plot(years,dolphins,label='Dolphins')
plot.plot(years,whales,label='Whales')
plot.xlabel('Years')
plot.ylabel('WildLife Population')
plot.title('WildLife Population Per Year')
plot.legend(loc='upper left')

plot.show()
plot.close()
file3data.close()
```
